# Using dataset

```python
# Taking the data to be used from the dataframe
type_data = pd.DataFrame({"Shipping Mode": df["Shipping Mode"],
                  "Order Country": df["Order Country"],
                  "Category Id": df["Category Id"],
                  "Order Item Quantity": df["Order Item Quantity"],
                  "Product Price": df["Product Price"],
                  "sales": df["Sales per customer"],
                  "scheduled": df["Days for shipment (scheduled)"],
                  "real" : df["Days for shipping (real)"],
                  "risk" :df["Late_delivery_risk"],
                  "Type" : df["Type"]})
type_data.head(5)
```

```python
# It is the function of changing object to integer.
def change_Type(x):
    # cid: changing id
    unique = np.unique(x)
    data = []
    for i in range(0,len(unique)):
        data.append(i)
    data2 = pd.DataFrame({"col1" :data, "col2":unique})
    cid = []
    for i in x:
        for j in data2.iloc[:, :].values:
            if(i == j[1]):
                cid.append(j[0])
    return cid
    
```

```python
# Converting Order Country object to CountryId int.
CountryId = change_Type(type_data["Order Country"])
type_data.insert(1, "CountryId", CountryId, True) 
```

```python
# Converting Shiipin Mode object to ModeId int.
ModeId = change_Type(type_data["Shipping Mode"])
type_data.insert(1, "ModeID", ModeId, True)
```

```python
# Droping object data in using dataset.
type_data.drop('Shipping Mode', inplace=True, axis=1)
type_data.drop('Order Country', inplace=True, axis=1)
```

```python
# It is the function of linear regression.
def fitting_predicting(x, y):
    # fitting and predicting using inliers data
    lin_regressor = LinearRegression()
    x_input = x[:,np.newaxis]
    lin_regressor.fit(x_input, y)

    # predicting 
    y_pred = lin_regressor.predict(x_input)
    return y_pred
```

```python
# finding upper_range of dataset
def outlier_treatment(datacolumn):
    sorted(datacolumn)
    Q1,Q3 = np.percentile(datacolumn , [25,75])
    IQR = (Q3 - Q1)
    lower_range = Q1 -(1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
# removing outliers 
def removing_outlier(type_data, x, y, indexes):
    old_x = x

    
    for i in indexes:
        type_data.drop(labels=None, axis=0, index=i, columns=None, level=None, inplace=True, errors='raise')
        type_data.reset_index(inplace = True, drop = True) 

        x = np.delete(x,i)
        y = np.delete(y,i)
    return x, y, old_x, type_data
```

```python
# plotting wanted datasets
def plotter(x_inliers, y_inliers,old_x, y_pred):
    plt.figure(figsize=(10, 5))
    plt.scatter(x_inliers, y_inliers, s = 5,color='blue')
    plt.title('Removing Outlier : Shipping Time')
    plt.scatter(old_x, y_pred, s = 5, color='green')
    plt.xlabel('Real Shipping Time', fontsize=16)
    plt.ylabel('Schuled Shipping Time', fontsize=16)
    return plt.show()
```

```python
# Finding and removing the outlier with the help of linear regression 
# that finds the actual order time from the planned order time.
x = type_data["scheduled"].values
y = type_data["real"].values

inliers_y = y
inliers_x = x

# fitting data using linear regression
y_pred = fitting_predicting(x, y)
flag = True
while(flag):
    # finding distance between predicting data and given data 
    residual = pd.DataFrame({'distance' :((y_pred - inliers_y) ** 2)})
    lower_range, upper_range = outlier_treatment(residual)
    # finding indexes all outliers
    indexes1 = np.where(residual.distance > upper_range)
    indexes2 = np.where(residual.distance < lower_range)
    indexes = indexes1 + indexes2
    print(len(indexes[0]))

    # checking there is any index for outliers
    if len(indexes[0]) < 1:
        # ploting inliers outliers and predicting with using inliers data
        print("There are not outliers")
        flag = False
        break
    else:
        flag = True
  
    # plotting Tukey Box
    residual.boxplot(return_type = "dict")
    plt.plot()
        
    # removing outliers
    inliers_x, inliers_y, old_x, type_data = removing_outlier(type_data, inliers_x, inliers_y, indexes)
    
    # fitting and predicting using inliers data
    y_pred = fitting_predicting(inliers_x, inliers_y)
    old_x = inliers_x
    
    # ploting inliers outliers and predicting with using inliers data
    plotter(inliers_x, inliers_y, inliers_x, y_pred)
```

# Pca

```python
# Converting Type to int
ModeId = change_Type(type_data["Type"])
type_data.insert(1, "TypeId", ModeId, True)
```

```python
# dropping object data in dataset.
type_data.drop('Type', inplace=True, axis=1)
```

```python
pca = PCA(n_components=3)
pca.fit(type_data)
x_pca = pca.transform(type_data)
```

```python
#creating dataframe from 2 dimensional data
df_pca = pd.DataFrame(data = x_pca
             , columns = ['pca1', 'pca2','pca3'])
```

```python
print(f"Before PCA: {type_data.shape}") 
print(f"After PCA: {df_pca.shape}")
```

```python
df_pca.head(10)
```

```python
ax = plt.figure(figsize=(12, 6)).gca(projection='3d')
ax.scatter(df_pca.pca1, df_pca.pca2, df_pca.pca3, color='green', alpha=1)
ax.set(xlabel='X1', ylabel='X2', zlabel='y')

plt.show()
```

```python
# appliying Kmeans cluster for dataset which appliying pca (303,2)
Kmean = KMeans(n_clusters=3)
Kmean.fit(df_pca)
center_pca = Kmean.cluster_centers_


print(f"After appliying PCA Kmeans cluster center: {center_pca}")
```

```python
ax = plt.figure(figsize=(18, 9)).gca(projection='3d')
ax.scatter(df_pca.pca1, df_pca.pca2, df_pca.pca3, color='green', alpha=1)
ax.scatter(center_pca[:,0], center_pca[:,1], center_pca[:,2], color='red', alpha=1)
ax.set(xlabel='X1', ylabel='X2', zlabel='y')

plt.show()
```

```python
#Predict clusters
y_predict = Kmean.fit_predict(df_pca)

# Creating new data frame for adding Cluster
new_df_pca = pd.DataFrame(data = df_pca, columns = ['pca1', 'pca2','pca3'])

# Adding a new column that Cluster
new_df_pca['Cluster'] = y_predict

# Change Cluster colums according the predictions
cluster1 = new_df_pca[new_df_pca.Cluster == 0]
cluster2 = new_df_pca[new_df_pca.Cluster == 1]
cluster3 = new_df_pca[new_df_pca.Cluster == 2]

new_df_pca.head()
```

```python
ax = plt.figure(figsize=(18, 9)).gca(projection='3d')
ax.scatter(cluster1.pca1, cluster1.pca2, cluster1.pca3, color='green', alpha=1)
ax.scatter(cluster2.pca1, cluster2.pca2, cluster2.pca3, color='blue', alpha=1)
ax.scatter(cluster3.pca1, cluster3.pca2, cluster3.pca3, color='orange', alpha=1)
ax.scatter(center_pca[:,0], center_pca[:,1], center_pca[:,2], color='red', alpha=1)
ax.set(xlabel='X1', ylabel='X2', zlabel='y')

plt.show()
```

# Text Mining

```python
comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in np.unique(df["Product Name"]): 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
```

```python
# Generate a word cloud image
mask = np.array(Image.open("source.png"))
wordcloud = WordCloud(stopwords=stopwords, background_color="white", mode="RGBA", max_words=1000, mask=mask).generate(comment_words)

# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[7,7])
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")

# store to file
plt.savefig("created.png", format="png")

plt.show()
```
