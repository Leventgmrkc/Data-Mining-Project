```python
basket = (df[df['Category Id']== 73]
         .groupby(['Order Id','Order City'])['Customer Zipcode']
         .sum().unstack().reset_index().fillna(0)         
         .set_index('Order Id'))
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
basket_sets
```
