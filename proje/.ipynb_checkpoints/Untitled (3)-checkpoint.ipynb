```python
import pandas as pd
import numpy as np
#df = pd.read_csv("DataCoSupplyChainDataset.csv", encoding ="latin-1") 
#df.head(50)

type_data = pd.read_csv("DataCoSupplyChainDataset.csv", encoding ="latin-1", 
                        usecols = ["Order Country","Type","Category Id",
                                   "Order Item Product Price","Order Item Quantity",
                                   "Product Category Id", "Product Price","Shipping Mode"]
)
type_data.head(5)
```

```python
x = type_data["Shipping Mode"]
#y = type_data["Shipping Mode"]
```

```python
deneme = np.unique(x)
data = []
for i in range(0,len(deneme)):
    data.append(i)
data = pd.DataFrame({'Mode' :data, 'Shipping Mode':deneme})
```

```python
cid = []
for i in x:
    for j in data.iloc[:, :].values:
        if(i == j[1]):
            cid.append(j[0])
```

```python
type_data.insert(1, "Mode", cid, True) 
```

```python
type_data.drop('Shipping Mode', inplace=True, axis=1)
type_data.head(10)
```

```python
new_col = pd.read_csv("DataCoSupplyChainDataset.csv", encoding ="latin-1",usecols = ["Sales"])
print((np.unique(new_col)))

type_data.insert(1, "deneme", new_col, True) 
#type_data.drop('order', inplace=True, axis=1)
```

```python
type_data.to_csv('odeme_type.csv')
```
